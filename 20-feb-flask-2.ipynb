Q1. Explain GET and POST methods.



In GET method we can not send large amount of data rather limited data is sent because the request parameter is appended into the URL. GET requests are the most common and widely used methods in APIs and websites. Simply put, the GET method is used to retreive data from a server at the specified resource. For example, say you have an API with a /users endpoint. Making a GET request to that endpoint should return a list of all available users

In POST method large amount of data can be sent because the request parameter is appended into the body.
POST is an HTTP method designed to send data to the server from an HTTP client. The HTTP POST method requests the web server accept the data enclosed in the body of the POST message. HTTP POST method is often used when submitting login or contact forms or uploading files and images to the server.

HTTP (Hypertext Transfer Protocol) defines a set of request methods or "verbs" to indicate the desired action to be HTTP performed for a given resource. Two of the most commonly used methods are GET and POST.

Here's a brief explanation of each:

GET: The GET method is used to retrieve information from a server. When you enter a URL in your web browser's address bar, it sends a GET request to the server, which then responds with the requested information. The GET method is also used for submitting form data, but the data is included in the URL as query parameters. Since the data is included in the URL, it is less secure than the POST method, which hides the data in the body of the request.

POST: The POST method is used to submit data to a server to create or update a resource. When you fill out a form on a website and click the submit button, the form data is sent to the server using a POST request. The data is included in the body of the request, which makes it more secure than the GET method, as the data is not visible in the URL. The POST method is also used for uploading files to a server.

In summary, GET is used to retrieve information, while POST is used to submit information to a server to create or update a resource. GET requests are less secure but can be cached by web browsers, while POST requests are more secure but cannot be cached

Q2. Why is request used in Flask?



The Request, in Flask, is an object that contains all the data sent from the Client to Server. This data can be recovered using the GET/POST Methods.The Request, in Flask, is an object that contains all the data sent from the Client to Server. This data can be recovered using the GET/POST Methods.

In Flask, the request object is used to access incoming request data. It allows you to access data submitted in the body of an HTTP request, as well as data submitted as query parameters in the URL.

Here are a few reasons why the request object is useful in Flask:

Handling form data: When you submit a form on a web page, the form data is sent to the server as part of the HTTP request. The request object allows you to access this data in your Flask application, so you can process it and take appropriate actions.

Retrieving request headers: HTTP requests can contain a variety of headers that provide additional information about the request, such as the user agent, accept-language, or authorization header. The request object allows you to access these headers in your Flask application.

Parsing JSON data: When sending data to a server, it is often sent as JSON format in the request body. Flask uses the request object to parse the JSON data and make it available for processing.

Working with cookies: HTTP requests can also contain cookies, which are small pieces of data stored on the client-side that can be accessed by the server. The request object allows you to access and manipulate cookies in your Flask application.
